# K-means clustering

## import library

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors as colors
from matplotlib import cm
```

## load data

```python
fname_data = 'assignment_11_data.csv'

feature = np.genfromtxt(fname_data, delimiter=',')

x = feature[:,0]
y = feature[:,1]

number_data     = np.size(feature, 0)
number_feature  = np.size(feature, 1)

print('number of data : {}'.format(number_data))
print('number of feature : {}'.format(number_feature))
```

```python
plt.figure(figsize=(8,8))   
plt.title('input data')

plt.scatter(x, y, color='blue')

plt.tight_layout()
plt.show()
```

## compute distance
- feature : $n \times m$, center : $1 \times m$, distance : $n \times 1$
- $n$ : number of data, $m$ : number of features

```python
def compute_distance(feature, center):

    # ++++++++++++++++++++++++++++++++++++++++++++++++++
    # complete the blanks
    #
    n=len(feature)
    distance=np.zeros((n))
    for i in range(n):
        distance[i] = np.linalg.norm(center-feature[i])

        
    #
    # ++++++++++++++++++++++++++++++++++++++++++++++++++

    return distance
```

## compute centroid
- feature : $n \times m$, label_feature : $n \times 1$, value_label : $1 \times 1$, centroid : $1 \times m$
- $n$ : number of data, $m$ : number of features

```python
def compute_centroid(feature, label_feature, label):

    # ++++++++++++++++++++++++++++++++++++++++++++++++++
    # complete the blanks
    #
    i=np.where(label_feature==label)
    centroid=np.mean(feature[i],axis=0)

    #
    # ++++++++++++++++++++++++++++++++++++++++++++++++++

    return centroid
```

## compute label
- distance : $n \times k$, label_feature : $n \times 1$
- $n$ : number of data, $k$ : number of clusters

```python
def compute_label(distance):

    # ++++++++++++++++++++++++++++++++++++++++++++++++++
    # complete the blanks
    #
    label_feature=np.argmin(distance,axis=1)

    #
    # ++++++++++++++++++++++++++++++++++++++++++++++++++
    
    return label_feature 
```

```python
def compute_loss(center,feature,label_feature,number_cluster):
    loss=0
    for i in range(number_cluster):
        index=np.where(label_feature==i)
        loss=loss+compute_distance(feature[index],center[i]).sum()
    loss=loss/number_data
    return loss
```

## the number of clusters $K = 2$

```python
number_cluster          = 2
number_iteration        = 30     # you can modify this value
loss_iteration_02       = np.zeros(number_iteration)
centroid_iteration_02   = np.zeros((number_iteration, number_cluster, number_feature))
label_feature_02        = np.random.randint(0, number_cluster, size=(number_data))
```

```python
# ++++++++++++++++++++++++++++++++++++++++++++++++++
# complete the blanks
#
distance=np.zeros((number_data,number_cluster))
centroid=np.zeros((number_cluster,number_feature))

index=np.random.randint(number_data,size=2)
centroid=feature[index,:]
label_cluster=np.arange(number_cluster)
for i in range(number_iteration):
    for k in range(len(label_cluster)):
        distance[:,k]=compute_distance(feature,centroid[k])
    label_feature_02=compute_label(distance)
    loss_iteration_02[i]=compute_loss(centroid,feature,label_feature_02,len(label_cluster))
    
    for s in range(len(label_cluster)):
        if np.any(label_feature_02==s): ## 교수님이 설명하신 centroid가 같을떄 구현
            centroid[s]=compute_centroid(feature,label_feature_02,s)
        else:
            centroid[s]=centroid[s]   
    centroid_iteration_02[i]=centroid
#
# ++++++++++++++++++++++++++++++++++++++++++++++++++
```
